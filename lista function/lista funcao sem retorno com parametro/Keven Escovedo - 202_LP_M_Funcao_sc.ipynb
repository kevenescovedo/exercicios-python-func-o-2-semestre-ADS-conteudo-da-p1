### 1. (Função sem retorno com parâmetro) Faça um programa contendo uma função/método que leia um número e o multiplique por 2, apresente o resultado.

```
def multiply_for_two(number):
    result = number * 2
    print(number,"multiplicado por 2 é: ", result)
  
def main() :
    number_input = float(input("Digite um número para que ele seja multiplicado por 2: "))
    multiply_for_two(number_input)
    

main()    
```

### 5. (Função sem retorno com parâmetro) Faça uma função/método para a partir de um preço de um produto informado, calcular e apresentar o novo preço reajustado em alguma porcentagem lida.

```
def ajuste_sal(preco,ajuste):
  ajuste_s = preco * (ajuste/100)
  preco_total = preco + ajuste_s
  print("o ajuste de",ajuste,"% do preço de",preco,"foi para",preco_total)
def main():
  preco = float(input("digite o preço do produto: "))
  ajuste = float(input("digite o ajuste que queria fazer em porcentagem: "))
  ajuste_sal(preco,ajuste)
main()  
```

### 6. (Função sem retorno com parâmetro) Faça uma função/método para a partir de um valor inicial e um valor final realizar o acumulo desse valores e apresentar o resultado.
    Exemplo:
        a = 2
        b = 8
        // 2 + 3 + 4 + 5 + 6 + 7 + 8 = 35
        r = 35

```
def acumulo(a,b):
  total = 0
  while a <= b:
    total = a + total
    a = a + 1
  print("valor acumulado (r):",total)
def main():
  a = int(input("digite o valor numero a: "))
  b = int(input("digite o valor numero b: "))
  if a <= b:
   acumulo(a,b)
  else:
    print("o a tem que ser menor que b") 
main()  

```

### 8. (Função sem retorno com parâmetro) Faça uma função/método para calcular a tabuada de um número informado pelo usuário.

```
def tabuada(n):
  for x in range(101):
    resultado = n * x
    print(n,"x", x, "=",resultado)
def main():
   n = int(input("digite um numero para ver sua tabuada: ")) 
   tabuada(n)
main()   

```

### 10. (Função sem retorno com parâmetro) Faça uma função/método para verificar e contar quantas letras a tem numa frase.

```
def contar(frase):
  algo = frase.replace(" ","")
  contador = len(algo)
  print("a frase", frase,"tem",contador,"letras")
def main():
  frase = input("Escreva uma frase: ")
  contar(frase)
main()   


```
